```python
from bs4 import BeautifulSoup
import requests

#####################################################
# Extracting the links of multiple movie transcripts
#####################################################

# Define the root URL
root = 'https://subslikescript.com'
website = f'{root}/movies_letter-A'
result = requests.get(website)
content = result.text
soup = BeautifulSoup(content, 'html.parser')

# Pagination
pagination = soup.find('ul', class_='pagination')
pages = pagination.find_all('li', class_='page-item')
last_page = pages[-2].text  # Get the second to last element, which is the last page number

links = []

# Loop through the first 3 pages (modify as needed)
for page in range(1, int(last_page) + 1)[:3]:
    page_url = f'{website}?page={page}'  # Fixing the page URL
    result = requests.get(page_url)
    content = result.text
    soup = BeautifulSoup(content, 'html.parser')

    # Locate the box that contains a list of movies
    box = soup.find('article', class_='main-article')

    # Store each link in the "links" list (concatenate root URL)
    for link in box.find_all('a', href=True):  # find_all returns a list
        links.append(link['href'])

#################################################
# Extracting the movie transcript
#################################################

# Loop through the "links" list and send a request to each link
for link in links:
    try:
        # Full link to the movie page
        movie_url = f'{root}/{link}'
        print(f'Processing: {movie_url}')
        
        result = requests.get(movie_url)
        content = result.text
        soup = BeautifulSoup(content, 'html.parser')

        # Locate the box that contains title and transcript
        box = soup.find('article', class_='main-article')

        # Safely locate the title
        title = box.find('h1').get_text() if box.find('h1') else 'No Title'

        # Safely locate the transcript
        transcript_div = box.find('div', class_='full-script')
        transcript = transcript_div.get_text(strip=True, separator=' ') if transcript_div else 'Transcript not found'

        # Export the transcript to a text file named after the movie title
        with open(f'{title}.txt', 'w', encoding='utf-8') as file:
            file.write(transcript)

    except Exception as e:
        print(f'----------- Error processing link: {movie_url}')
        print(f'Error details: {e}')
```
